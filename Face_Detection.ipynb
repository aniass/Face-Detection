## Face Detection using OpenCV

In this project was used OpenCV library to detect faces in an image. Face detection is performed by using classifiers. The OpenCV library has two pre-trained face detection classifiers which are ready to use. In this task was used Harr Classifier  with the detectMultiscale module.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the image

face_img = cv2.imread('People1.jpg')
face_img2 = cv2.cvtColor(face_img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(face_img, cv2.COLOR_BGR2GRAY)

plt.title("Original image")
plt.imshow(face_img2)
plt.axis('off')
plt.show()

plt.title('Grayscale image')
plt.axis('off')
plt.imshow(gray_img, cmap='gray')
plt.show()
```

The function that converts the BGR image format to RBG to display the true colors of the image:

```python
def convertColor(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Loading the classifier for frontal face
facePath = 'C:\\Users\\PC\\Anaconda3\\envs\\opencv\\Library\\etc\\haarcascades\\haarcascade_frontalface_default.xml'

# Create the Harr Cascade
face_cascade = cv2.CascadeClassifier(facePath)

# Detect faces in the image
faces_detect = face_cascade.detectMultiScale(
              gray_img,
              scaleFactor=1.8,
              minNeighbors=5,
              minSize=(0, 0),
              flags = cv2.CASCADE_SCALE_IMAGE
              )

# Draw a rectangle around the each faces
for (x, y, w, h) in faces_detect:
    face_img = cv2.rectangle(face_img, (x, y), (x+w, y+h), (0, 255, 0), 9)
    roi_gray = gray_img[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]

print('Found {0} faces!'.format(len(faces_detect)))

plt.imshow(convertColor(face_img))
```
